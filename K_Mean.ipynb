```python
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=500, centers=3, n_features=2,cluster_std=1.9)
X
```

```python
plt.scatter(X[:,0], X[:,1], c=y,cmap="autumn")
```

```python
idx = np.random.permutation(np.arange(X.shape[0]))
idx.size
```

```python
# M0 = X[idx[0],:]
# M1 = X[idx[1],:]
# M2 = X[idx[2],:]
M0 = X[y==0].mean(axis=0)
M1 = X[y==1].mean(axis=0)
M2 = X[y==2].mean(axis=0)
print(M0, M1, M2)
```

```python
def distance(x,y):
    return np.sqrt(np.sum((x-y)**2,axis=1))
```

```python
def group_mean(X,M0,M1,M2):
    d1 = distance(X,M0)
    d2 = distance(X,M1)
    d3 = distance(X,M2)
    d = np.hstack((d1[:,np.newaxis],d2[:,np.newaxis],d3[:,np.newaxis]))
    return np.argmin(d,axis=1)
```

```python
y_new = group_mean(X,M0,M1,M2)
```

```python
plt.scatter(X[:,0], X[:,1], c=y_new,cmap="autumn")
```
