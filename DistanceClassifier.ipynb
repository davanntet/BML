```python
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=150,centers=2)
```

```python
y[y==0]=-1
np.unique(y)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap="autumn")
```

```python
X1 = X[y == 1]
X2 = X[y == -1]
print(X1.shape,X2.shape)
```

```python
M1 = X1.mean(axis=0)
M2 = X2.mean(axis=0)
```

```python
idx = np.random.randint(y.size)
r = X[idx,:]
d1 = ((X[idx]-M1)**2).sum()
d2 = ((X[idx]-M2)**2).sum()
if d1<d2:
    print(y[idx],1)
else:
    print(y[idx],-1)
plt.scatter((X[idx,0]-M1[0])**2,(X[idx,1]-M1[1])**2,c='g',cmap="autumn")
plt.scatter((X[idx,0]-M2[0])**2,(X[idx,1]-M2[1])**2,c='b',cmap="autumn")
plt.scatter(M1[0],M1[1],c="g",marker="x",s=100)
plt.scatter(M2[0],M2[1],c="b",marker="x",s=100)
```
