```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(20)
y = 4*x +5 + np.random.randn(x.size)*3
plt.scatter(x,y)
```

# Formula for Linear Regression: Xa = Y
## we need to find a

```python
X1 = np.ones((x.size,1))
X = np.hstack((x[:,np.newaxis],X1))
X #now we have X
```

```python
Y = y[:,np.newaxis] # now we have Y
```

```python
a = np.linalg.lstsq(X,Y,rcond=None)
a
```

```python
y_predict = a[0][0]*x + a[0][1]
plt.scatter(x,y)
plt.plot(x,y_predict,'r')
```
